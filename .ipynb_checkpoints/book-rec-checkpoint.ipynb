```python
def predict_ratings(df, user_similarity):
    df_pred = df.copy()

    def predict_for_item(column):
        nan_indices = column[column.isna()].index  # User IDs where ratings are NaN

        for user_id in nan_indices:
            if user_id not in user_similarity.index:
                continue  # Skip if user ID is not in similarity matrix

            sim_scores = user_similarity.loc[user_id]  # Similarity scores for this user
            rated_users = column.dropna().index  # Users who rated this item

            # Get top 5 most similar users who rated this item
            top_users = sim_scores.loc[rated_users].nlargest(5)

            if top_users.sum() > 0:  # Avoid division by zero
                pred_rating = np.dot(top_users, column.loc[top_users.index]) / top_users.sum()
            else:
                pred_rating = 0  # Default if no valid ratings available

            df_pred.at[user_id, column.name] = pred_rating  # Restore user mean

    df.apply(predict_for_item, axis=0)  # Apply column-wise
    return df_pred
```

## User Based

```python
df_filled_user = df_filled.T
```

```python
model_knn = NearestNeighbors(metric = 'cosine', algorithm = 'brute')
model_knn.fit(df_filled_user)
```

```python
# Pick a random user index
query_index = np.random.choice(df_filled_user.shape[0])

# Find nearest neighbors (similar users)
distances, indices = model_knn.kneighbors(df_filled_user.iloc[query_index, :].values.reshape(1, -1), n_neighbors=6)

# Get the actual user ID
user_id = df_filled_user.index[query_index]

# Display similar users
print(f"Users similar to {user_id}:\n")
for i in range(1, len(indices.flatten())):  # Skip the first one (self)
    print(f"{i}: {df_filled_user.index[indices.flatten()[i]]}, Similarity: {1 - distances.flatten()[i]:.4f}")
```

```python
# Get similar users
similar_users = indices.flatten()[1:]  # Skip the first one (self)

# Get books rated by similar users
recommended_books = df_filled_user.iloc[similar_users].mean(axis=0).sort_values(ascending=False).index

print(f"\nRecommended books for User {user_id}:\n", list(recommended_books[:5]))  # Top 5 recommendations
```

## Model Based

```python
from scipy.sparse.linalg import svds
```

```python
# Convert DataFrame to NumPy array (ensure it's numeric)
df_matrix = df_filled_user.to_numpy(dtype=np.float64)  

# Perform SVD
U, sigma, Vt = svds(df_matrix, k=50)

# Convert sigma into a diagonal matrix
sigma = np.diag(sigma)

# Reconstruct the predicted ratings matrix
predicted_ratings = np.dot(np.dot(U, sigma), Vt)

# Convert back to DataFrame with original index & columns
predicted_ratings_df = pd.DataFrame(predicted_ratings, index=df_filled_user.index, columns=df_filled_user.columns)
```

```python
user_id = 254  # Replace with a real User-ID

# Get user's predicted ratings
user_ratings = predicted_ratings_df.loc[user_id]

# Sort books by highest predicted rating
recommended_books = user_ratings.sort_values(ascending=False).index

print(f"Recommended books for User {user_id}:\n", list(recommended_books[:5]))  # Top 5 recommendations
```
